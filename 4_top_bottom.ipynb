```python
# Load the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load your dataset
relative_path = os.path.join("data", "raw", "retail_sales_dataset.csv")
df = pd.read_csv(relative_path)
```

* Bar chart showing the total sales for three product categories: Electronics, Clothing, and Beauty. 

* Each bar's height represents the total sales amount, with the exact values displayed above each bar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales per product category
categories = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

# Visualize
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=categories.index, y=categories.values, hue=categories.index, palette='viridis', legend=False) # Fixed line

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', fontsize=10, color='black')

plt.title("Product Categories by Total Sales", fontsize=16)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Product Categories by Total Sales
category_sales = df.groupby('Product Category')['Total Amount'].sum()
categories = category_sales.sort_values(ascending=False)

print("Product Categories by Total Sales:\n", categories)

# Customers by Total Spending
customer_spending = df.groupby('Customer ID')['Total Amount'].sum()
customers = customer_spending.sort_values(ascending=False)

print("\Customers by Total Spending:\n", customers)

# Items by quantity sold.
item_quantity = df.groupby("Product Category")['Quantity'].sum()
items = item_quantity.sort_values(ascending = False)

print("\nItems by Quantity Sold:\n",items)
```
