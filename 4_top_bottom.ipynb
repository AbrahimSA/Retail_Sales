```python
# Load the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Load your dataset
relative_path = os.path.join("data", "retail_sales_dataset.csv")
df = pd.read_csv(relative_path)
```

```python
 
# 1. Top/Bottom Product Categories by Total Sales
category_sales = df.groupby('Product Category')['Total Amount'].sum()
top_categories = category_sales.sort_values(ascending=False).head(5)
bottom_categories = category_sales.sort_values(ascending=True).head(5)

print("Top 5 Product Categories by Total Sales:\n", top_categories)
print("\nBottom 5 Product Categories by Total Sales:\n", bottom_categories)

# Visualize
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar')
plt.title("Top 5 Product Categories by Total Sales")
plt.show()

plt.figure(figsize=(10, 6))
bottom_categories.plot(kind='bar')
plt.title("Bottom 5 Product Categories by Total Sales")
plt.show()

# 2. Top/Bottom Customers by Total Spending
customer_spending = df.groupby('Customer ID')['Total Amount'].sum()
top_customers = customer_spending.sort_values(ascending=False).head(5)
bottom_customers = customer_spending.sort_values(ascending=True).head(5)

print("\nTop 5 Customers by Total Spending:\n", top_customers)
print("\nBottom 5 Customers by Total Spending:\n", bottom_customers)

# 3. Top/Bottom Items by quantity sold.
item_quantity = df.groupby("Product Category")['Quantity'].sum()
top_items = item_quantity.sort_values(ascending = False).head(5)
bottom_items = item_quantity.sort_values(ascending = True).head(5)

print("\nTop 5 Items by Quantity Sold:\n",top_items)
print("\nBottom 5 Items by Quantity Sold:\n", bottom_items)
```
